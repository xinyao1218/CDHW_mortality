```python
from osgeo import gdal,osr,ogr
import cartopy.crs as ccrs
import numpy as np
import  matplotlib.pyplot as plt
import xarray as xr
import netCDF4 as nc
import scipy as sp
import geopandas as gpd
import pandas as pd
import glob
import rasterio
import rioxarray
import glob
```

```python
file_paths = glob.glob("C:\\Users\yaoxin\Desktop\ISIMIP3b\*.nc")
# 合并数据集
combined_data = xr.Dataset()

for file_path in file_paths:
    ds = xr.open_dataset(file_path)
    ds = ds.sel(
    lat=slice(54, 3),
    lon=slice(73, 136)
)
    combined_data = xr.merge([combined_data, ds])
    ds.close()

ds = None

monthly_mean = combined_data.resample(time='M').sum(dim='time')
combined_data = None
monthly_mean.to_netcdf('C:\\Users\yaoxin\Desktop\ISIMIP3b\com_pre_month\\ukesm1-0-ll_r1i1p1f2_w5e5_ssp585_pr_global_daily_2015_2100.nc')
```
