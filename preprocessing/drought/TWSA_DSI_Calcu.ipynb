```python
#读取数据，根据时间对中国数据进行切片
tws_his=xr.open_dataset('cwatm_gfdl-esm4_w5e5_historical_histsoc_default_tws_global_monthly_1850_2014.nc')
ssp126=xr.open_dataset('cwatm_gfdl-esm4_w5e5_ssp126_2015soc-from-histsoc_default_tws_global_monthly_2015_2100.nc')
ssp370=xr.open_dataset('cwatm_gfdl-esm4_w5e5_ssp370_2015soc-from-histsoc_default_tws_global_monthly_2015_2100.nc')
ssp585=xr.open_dataset('cwatm_gfdl-esm4_w5e5_ssp585_2015soc-from-histsoc_default_tws_global_monthly_2015_2100.nc')



tws_his = tws_his.sel(
    time=slice('1941-01-01', '2014-12-01'),
    lat=slice(54, 3),
    lon=slice(73, 136)
)

ssp126 = ssp126.sel(
    lat=slice(54, 3),
    lon=slice(73, 136)
)

ssp370 = ssp370.sel(
    lat=slice(54, 3),
    lon=slice(73, 136)
)

ssp585 = ssp585.sel(
    lat=slice(54, 3),
    lon=slice(73, 136)
)

time_base = tws_his.sel(
    time=slice('2004-01-01', '2009-12-01')
)

base_monthly_mean = time_base.groupby('time.month').mean('time')


twsa_his = tws_his.groupby('time.month')-base_monthly_mean
twsa_ssp126 = ssp126.groupby('time.month')-base_monthly_mean
twsa_ssp370 = ssp370.groupby('time.month')-base_monthly_mean
twsa_ssp585 = ssp585.groupby('time.month')-base_monthly_mean

tws_his = None
ssp126 = None
ssp370 = None
ssp585 = None


twsa_monthly_mean = twsa_his.groupby('time.month').mean('time')
twsa_monthly_std = twsa_his.groupby('time.month').std('time')

twsa_dsi_his = twsa_his.groupby('time.month') - twsa_monthly_mean
twsa_dsi_his = twsa_dsi_his.groupby('time.month') / twsa_monthly_std

twsa_dsi_ssp126 = twsa_ssp126.groupby('time.month') - twsa_monthly_mean
twsa_dsi_ssp126 = twsa_dsi_ssp126.groupby('time.month') / twsa_monthly_std

twsa_dsi_ssp370 = twsa_ssp370.groupby('time.month') - twsa_monthly_mean
twsa_dsi_ssp370 = twsa_dsi_ssp370.groupby('time.month') / twsa_monthly_std

twsa_dsi_ssp585 = twsa_ssp585.groupby('time.month') - twsa_monthly_mean
twsa_dsi_ssp585 = twsa_dsi_ssp585.groupby('time.month') / twsa_monthly_std

twsa_monthly_mean = None
twsa_monthly_std = None
twsa_his = None
twsa_ssp126 = None
twsa_ssp370 = None
twsa_ssp585 = None

combine_ssp126 = xr.concat([twsa_dsi_his, twsa_dsi_ssp126], dim='time')
combine_ssp370 = xr.concat([twsa_dsi_his, twsa_dsi_ssp370], dim='time')
combine_ssp585 = xr.concat([twsa_dsi_his, twsa_dsi_ssp585], dim='time')

twsa_dsi_his = None
twsa_dsi_ssp126 = None
twsa_dsi_ssp370 = None
twsa_dsi_ssp585 = None

combine_ssp126.to_netcdf('C:\\Users\Lenovo\Desktop\ISIMIP3b\TWSA_DSI\cwatm_gfdl-esm4_w5e5_ssp126_twsa_dsi.nc')

combine_ssp370.to_netcdf('C:\\Users\Lenovo\Desktop\ISIMIP3b\TWSA_DSI\cwatm_gfdl-esm4_w5e5_ssp370_twsa_dsi.nc')

combine_ssp585.to_netcdf('C:\\Users\Lenovo\Desktop\ISIMIP3b\TWSA_DSI\cwatm_gfdl-esm4_w5e5_ssp585_twsa_dsi.nc')

combine_ssp126 = None
combine_ssp370 = None
combine_ssp585 = None
```
